# Data Preprocessing

```python
import pandas as pd
```

```python
df = pd.read_csv("../data/hand_landmarks_dataset.csv")
```

```python
df
```

```python
import numpy as np
import ast
```

```python
# Normalize relative to wrist position (x0, y0)
normalized_df = df.copy()

for idx, row in df.iterrows():
    wrist_x = row["x0"]  # Wrist x coordinate
    wrist_y = row["y0"]  # Wrist y coordinate

    # Normalize all x and y coordinates relative to wrist
    for i in range(21):  # Assuming 21 landmarks (0-20)
        x_col = f"x{i}"
        y_col = f"y{i}"

        if x_col in df.columns and y_col in df.columns:
            normalized_df.loc[idx, x_col] = row[x_col] - wrist_x
            normalized_df.loc[idx, y_col] = row[y_col] - wrist_y
```

```python
normalized_df.to_csv(
    "../data/hand_landmarks_dataset_normalized_to_the_wrist.csv", index=False
)
```
